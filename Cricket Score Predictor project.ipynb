```python
# Importing Necessary libraries

import numpy as np
import pandas as pd
```

```python
# Loading the dataset

df = pd.read_csv('t20i_info.csv')
```

```python
df.head()
```

```python
# Checking for null values

df.isnull().sum()
```

```python
df.shape
```

# What all things we need to predict the score

batting team,
bowling team,
city, 
current score, 
balls left,
current run rate, 
last five/six ---> powerplay

```python
df[df['city'].isnull()]['venue'][0].split(" ")[0]
```

```python
# To fill null values

df['city'] = df['city'].fillna(df['venue'].apply(lambda x:x.split(' ')[0]))
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['city'].value_counts()
```

```python
# shortlisting cities which has 600+ matches on ground

eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
eligible_cities
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df.head()
```

```python
df.shape
```

```python
df.tail()
```

```python
# To Calculate the current score 


df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df['current_score']
```

```python
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# To add new columns in the data set

df['over'] = df['ball'].apply(lambda x : str(x).split(".")[0])
df['ball_no.'] = df['ball'].apply(lambda x : str(x).split(".")[1])
```

```python
df.head()
```

```python
# Numbers of balls bowled

df['balls_bowled'] = df['over'].astype(int)*6 + df['ball_no.'].astype(int)
```

```python
df.tail()
```

```python
# To check number of balls left in the match 

df['balls_left'] = 120 - df['balls_bowled']
```

```python
df.tail()
```

```python
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df.tail()
```

```python
# To check how many wickets are left 

df['player_dismissed'] = df['player_dismissed'].apply(lambda x:1 if x!='0' else '0')
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].astype(int)
```

```python
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
df.tail()
```

```python
df['wicket_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
# To find Current Run-rate

df['current_run_rate'] = (df['current_score']*6) / df['balls_bowled']
```

```python
df.tail()
```

```python
# To get runs of last five/six overs

groups = df.groupby('match_id')
match_id = df['match_id'].unique()
last_five = []
for id in match_id:
    last_five.extend(groups.get_group(id).rolling(window = 30)['runs'].sum().values.tolist())
```

```python
df['last_five'] = last_five
```

```python
last_five
```

```python
# Merging the dataset

final_df = df.groupby('match_id')['runs'].sum().reset_index().merge(df, on ='match_id')
```

```python
final_df
```

```python
df.isnull().sum()
```

```python
final_df.columns
```

```python
final_df = final_df[['batting_team', 'bowling_team','city', 'current_score','balls_left', 'wicket_left',
       'current_run_rate', 'last_five','runs_x']]
```

```python
final_df
```

```python
final_df.dropna(inplace = True)
```

```python
final_df
```

```python
final_df.isnull().sum()
```

# Training the dataset

```python
X = final_df.drop(columns = ['runs_x'])
Y = final_df['runs_x']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from xgboost import XGBRegressor
```

```python
pip install --upgrade scikit-learn
```

```python
transformer = ColumnTransformer([
    ('transformer', OneHotEncoder(sparse_output = False, drop = 'first'),['batting_team', 'bowling_team', 'city'])
], remainder = 'passthrough')
```

```python
pipe = Pipeline(steps = [
    ('step1' , transformer),
    ('step2', StandardScaler()),
    ('step3', XGBRegressor(n_estimators = 1000, learning_rate = 0.2, max_depth = 12, random_state = 1))
])
```

```python
pipe
```

```python
pipe.fit(X_train, Y_train)
Y_pred = pipe.predict(X_test)
```

```python
r2_score(Y_test, Y_pred)
```

```python
mean_absolute_error(Y_test, Y_pred)
```
